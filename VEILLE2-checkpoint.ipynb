<center><h1><font color='blue'>REGRESSION LOGISTIQUE</font><h1><center>

```python
data.head()
```

```python
data.isna().any()
```

```python
# encodage des variables catégorielles
for col in data_char.columns:
    le1 = LabelEncoder()
    data_char[col] = le1.fit_transform(data[col])
```

```python
data_char.head()
```

```python
data_num = data_quanti.drop(['pic'], axis = 1)
```

```python
data_num.head()
```

```python
stand = StandardScaler()
```

```python
data_num_stand = pd.DataFrame(stand.fit_transform(data_num), columns = data_num.columns)
```

```python
data_num_stand.head()
```

```python
df = pd.concat([data_num_stand, data_char], axis = 1)
```

```python
df
```

```python
df['pic'] = data['pic']
```

```python
df.head()
```

```python
y = df['coeur']
X = df.drop(['coeur'], axis = 1)
```

```python
X
```

## entrainement du modèle

```python
# partage du dataset
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(X_test.shape)
```

```python
# entrainement du modèle de régression logistique
modele1 = LogisticRegression()
modele1.fit(X_train, y_train)
```

```python
# performance du modèle1
modele1.score(X_test, y_test)
```

```python
# entrainement du modèle avec le bagging
modele2 = BaggingClassifier(n_estimators=130)
modele2.fit(X_train, y_train)
```

```python
# performance du modèle2
modele2.score(X_test, y_test)
```

```python
y_pred1 = modele1.predict(X_test)
y_pred2 = modele2.predict(X_test)
```

```python
confusion_logis = confusion_matrix(y_test, y_pred1)
```

```python
confusion_bagg = confusion_matrix(y_test, y_pred2)
```

```python
confusion_logis, confusion_bagg
```

```python
print(classification_report(y_test, y_pred1))
```

```python
print(classification_report(y_test, y_pred2))
```
